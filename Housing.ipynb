## Train test split

```python
#Splitting the data into training and testing data
import numpy as np
def split_train_test(data, test_ratio):
    np.random.seed(42)#This is a number given to this code so that it produces same results every time
    shuffled_indices=np.random.permutation(len(data))#shuffeling data indices
    test_set_size=int(len(data)*test_ratio)
    test_indices=shuffled[:test_set_ratio]
    train_indices=shuffled[test_set_ratio:]
    return data.iloc[train_indices], data.iloc[test_indices]

#For the above code sklearn provides a function that does the same work with some additional featuers    
from sklearn.model_selection import train_test_split
train_set, test_set= train_test_split(housing, test_size=0.2, random_state=42)
```

```python
# if there is a important feature on which my result is heavily dependent then it should be equally divided in the training and 
# testing set, because if the whole will be in testing then the model will not be trained according to that situtation. So it
# should be equally divided in them. In this case We use stratified sampling.

# We are making this income_cat because most of the median income values are clustered around 1.5 and 6 but some go far from 6
# . It is important to have a sufficient number of instances in your dataset for each stratum, or else the estimate
#  of the stratums importance may be biased. This means that you should not have too  many strata, and each stratum should be large enough.
#
# many strata, and each stratum should be large enough.
housing["income_cat"]=pd.cut(housing["median_income"], bins=[0, 1.5, 3.0, 4.5, 6.0, np.inf], labels=[1, 2, 3, 4, 5])

from sklearn.model_selection import StratifiedShuffleSplit
split=StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(housing, housing["income_cat"]):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
    
```

```python
#Now we will remove the income_cat attribute
for set_ in (strat_train_set, strat_test_set):
    set_.drop("income_cat", axis=1, inplace=True)
    
```
